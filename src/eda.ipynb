```python
import pandas as pd
import numpy as np
```

```python
cases_train = pd.read_csv('../data/cases_2021_train.csv')
cases_test = pd.read_csv('../data/cases_2021_test.csv')
```

```python
hist_lat_long = cases_train[["latitude", "longitude"]].plot.hist(bins=60)
```

```python
hist_outcome_group = cases_train['outcome_group'].value_counts().plot(kind='bar')
```

```python
hist_country_first_five = cases_train['country'].value_counts().iloc[0:5].plot(kind='bar', logy=True)
```

```python
hist_country_all = cases_train['country'].value_counts().plot(kind='bar', logy=True)
```

```python
hist_lat_long.get_figure().savefig('../plots/task-1.3/hist_lat_long.png')
hist_outcome_group.get_figure().savefig('../plots/task-1.3/hist_outcome_group.png')
hist_country_first_five.get_figure().savefig('../plots/task-1.3/hist_country_first_five.png')
hist_country_all.get_figure().savefig('../plots/task-1.3/hist_country_all.png')
```
